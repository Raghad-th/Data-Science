```python
import math
import pandas as pd
import sklearn
from sklearn import tree, neighbors, neural_network, svm
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
# load data
data = pd.read_csv ('antenna.csv')
```

```python
data.shape
```

```python
colName = 'GHI'
input = data.drop(colName, axis=1).values
output = data[colName].values
```

```python
# split data (training = 80, testing = 20)
x_train, x_test, y_train, y_test = train_test_split(input, output, shuffle=False, test_size=365)
```

```python

# get the model name from Scikit Learn
model = neural_network.MLPRegressor()
#model = svm.SVR()
#model = tree.DecisionTreeRegressor()
#model = RandomForestRegressor()
#model = neighbors.KNeighborsRegressor(n_neighbors=20)
# train the model
model.fit(x_train, y_train)
```

```python
# prediction
y_estimated = model.predict(x_test)

# score, metrics (mse, rmse, mae, mbe, score, R)
mse = mean_squared_error(y_test, y_estimated)
rmse = math.sqrt(mse)
mae = mean_absolute_error(y_test, y_estimated)
mbe = sum([(y_test[i] - y_estimated[i]) for i in range(0, len(y_test))]) / len(y_test)
m_test = sum(y_test) / len(y_test)
m_estimated = sum(y_estimated) / len(y_estimated)
up = sum([(y_test[i] - m_test) * (y_estimated[i] - m_estimated) for i in range(0, len(y_test))])
sum1 = sum([pow(y_test[i] - m_test, 2) for i in range(0, len(y_test))])
sum2 = sum([pow(y_estimated[i] - m_estimated, 2) for i in range(0, len(y_test))])
down = math.sqrt(sum1 * sum2)
r = up / down
#score = model.score(x_test, y_test)

print("mse = %s" % mse)
print("rmse = %s" % rmse)
print("mae = %s" % mae)
print("mbe = %s" % mbe)
print("r = %s" % r)
#print("score = %s" % score)

plt.plot(y_test, color='green', label='original')
plt.plot(y_estimated, color='red', label='estimated')
#plt.show()
```
